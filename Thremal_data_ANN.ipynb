```python
import pandas as pd
import numpy as np
```

```python
df.rename(columns={"AT":"Average_Temperature",
                   'V':"Exhuast_Vacuume",
                   "AP":"Ambient_Pressure",
                   'RH':"Relative_Humidity",
                   'PE':"Net_Hourly_Electrical_Energy_Consumption"},inplace=True)
```

```python
df.corr()
```

```python
# SPlit the data bet Independant Features and dependant features
X = df.drop('Net_Hourly_Electrical_Energy_Consumption',axis=1).values
X
```

```python
y= df['Net_Hourly_Electrical_Energy_Consumption'].values
y
```

```python
# split the data into train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.80,random_state=1)
X_train
```

```python
y_train
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
#model architecture
df.shape
```

```python
import tensorflow as tf
from  tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#hidden layers
model = Sequential()
model.add(tf.keras.layers.Input(shape=[4], name="Input_Layer"))
model.add(Dense(units=16, activation='relu', name= "Hidden_Layer_1"))
model.add(Dense(units=8, activation='elu', name = "Hidden_layer_2"))
model.add(Dense(units=1, name = "output_layer"))
```

```python
model.compile(
    loss = "mean_squared_error",
    optimizer = "adam"
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
 model.fit(
    x = X_train_scaled,
    y = y_train,
    epochs = 100,
    batch_size = 32,
    validation_data = (X_test_scaled,y_test),
    callbacks = EarlyStopping(monitor="val_loss",patience=10)
)
```

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)
```

```python
model.evaluate(X_train_scaled, y_train)
```

```python

model.evaluate(X_test_scaled, y_test)
```

```python
prediction = model.predict(X_test_scaled)
prediction
```

```python
prediction_df = pd.DataFrame(np.ravel(prediction),columns = ['predictions'])
prediction_df
```

```python
compare_df = pd.concat([pd.DataFrame(y_test,columns=['true_values']),prediction_df],axis=1)
compare_df
```
