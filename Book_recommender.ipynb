```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('/content/books.csv')
book
```

```python
book.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(book.isna().transpose(), cbar=False, ax=ax )

plt.xlabel('Columns')
plt.ylabel('Missing values')

plt.show ()
```

```python
#finding missing values
book["missing_description"] = np.where(book["description"].isna(), 1, 0)
book["age_of_book"]= 2024 -book["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
correlation_matrix = book[columns_of_interest].corr(method ="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot= True, fmt=".2f", cmap="coolwarm", cbar_kws={"label":"spearman correlation"})

heatmap.set_title("correlation heatmap")
plt.show()
```

```python

book_missing = book[~(book['description'].isna())&
    ~( book['num_pages'].isna())&
     ~(book['average_rating'].isna())&
    ~(book['published_year'].isna())
     ]

book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
#filtering code
book_missing_25_words = book_missing[book_missing["words_in_description"]>=25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(" :". join, axis=1)
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"]= book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("book_missing_25_words.csv", index=False)
    )
```
